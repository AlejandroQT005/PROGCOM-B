```python
# Ejercicio 1
class Animal:
    def mostrar_tipo(self):
        print("Este es un animal")

class Perro(Animal):
    pass

perro = Perro()
perro.mostrar_tipo()


# Ejercicio 2
class Persona:
    def __init__(self, nombre):
        self.nombre = nombre

class Estudiante(Persona):
    pass

estudiante = Estudiante("Carlos")
print(estudiante.nombre)


# Ejercicio 3
class Vehiculo:
    def arrancar(self):
        print("Vehículo encendido")

class Moto(Vehiculo):
    def arrancar(self):
        print("Moto encendida")

moto = Moto()
moto.arrancar()


# Ejercicio 4
class Empleado:
    def trabajar(self):
        print("Empleado trabajando")

class Gerente(Empleado):
    def trabajar(self):
        print("Gerente gestionando")

class Obrero(Empleado):
    def trabajar(self):
        print("Obrero construyendo")

gerente = Gerente()
obrero = Obrero()
gerente.trabajar()
obrero.trabajar()


# Ejercicio 5
class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

class Libro(Producto):
    pass

libro = Libro("El Principito", 15.5)
print(f"Libro: {libro.nombre}, Precio: ${libro.precio}")


# Ejercicio 6
class Persona:
    def saludar(self):
        print("Hola")

class Profesor(Persona):
    def enseñar(self):
        print("Estoy enseñando")

profesor = Profesor()
profesor.saludar()
profesor.enseñar()


# Ejercicio 7
class Animal:
    def sonido(self):
        print("Sonido genérico")

class Gato(Animal):
    def sonido(self):
        super().sonido()
        print("Miau")

gato = Gato()
gato.sonido()


# Ejercicio 8
class Computadora:
    def encender(self):
        print("Computadora encendida")

class Portatil(Computadora):
    pass

portatil = Portatil()
portatil.encender()


# Ejercicio 9
class Cuenta:
    def __init__(self, saldo):
        self.saldo = saldo

class CuentaAhorro(Cuenta):
    pass

cuenta_ahorro = CuentaAhorro(1000)
print(f"Saldo: ${cuenta_ahorro.saldo}")


# Ejercicio 10
class SerVivo:
    def respirar(self):
        print("Respirando...")

class Humano(SerVivo):
    pass

humano = Humano()
humano.respirar()
```
